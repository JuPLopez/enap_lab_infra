### 2. Resposta: recursos liberados

```python
import pandas as pd
df_resposta_recurso = pd.read_excel("dados/Relat√≥rio Gerencial de Resposta - Recursos Liberados1.xlsx", 
                               skiprows=2, 
                dtype={
                    'IBGE': str}
                        )

df_resposta_recurso = df_resposta_recurso.rename(columns={'IBGE': 'ibge7'})
df_resposta_recurso.head()
```

```python
# Converter as colunas para datetime
df_resposta_recurso['Data Empenho'] = pd.to_datetime(df_resposta_recurso['Data Empenho'], format='%d/%m/%Y', errors='coerce')
df_resposta_recurso['Data Envio Processo'] = pd.to_datetime(df_resposta_recurso['Data Envio Processo'], format='%d/%m/%Y', errors='coerce')

# Calcular a diferen√ßa
df_resposta_recurso['tempo_medio'] = df_resposta_recurso['Data Empenho'] - df_resposta_recurso['Data Envio Processo']

df_resposta_recurso[['Data Envio Processo', 'Data Empenho', 'tempo_medio']].head()
```

```python
top_10 = df_resposta_recurso.nsmallest(10, 'tempo_medio')
top_10
```

#### BASE Restabelecimento tempo m√©dio e contagem (por munic√≠pio)

```python
df_restabelecimento_mun = df_resposta_recurso.pivot_table(
    index='ibge7',           # Munic√≠pios como linhas
#    columns='tempo_medio',     
    values='tempo_medio',      # Qualquer coluna para contar (pode ser 'ibge7' tamb√©m)
    aggfunc=['mean', 'count'],         # Fun√ß√£o de agrega√ß√£o: contar ocorr√™ncias
    fill_value=0             # Preencher com 0 onde n√£o h√° ocorr√™ncias
).reset_index()

df_restabelecimento_mun.head()

df_restabelecimento_mun = df_resposta_recurso.groupby('ibge7').agg({
    'tempo_medio': ['mean', 'count']
}).reset_index()

# Ajustar nomes das colunas
df_restabelecimento_mun.columns = [
    'ibge7', 
    'tempo_medio_restabelecimento', 'qtd_restabelecimento'
]

df_restabelecimento_mun.head()
```

### 3. Reconstru√ß√£o: recursos liberados

```python
import pandas as pd
df_reconstrucao_recurso = pd.read_excel("dados/Relat√≥rio Gerencial de Reconstru√ß√£o - Recursos Liberados1.xlsx", 
                               skiprows=3, 
                                #skipfooter=1,
                dtype={
                    'IBGE': str}
                        )
#df_reconstrucao_recurso = df_reconstrucao_recurso.drop(['CNPJ','UF','Munic√≠pio', 'N¬∫ Siafi','N¬∫ munic√≠pio no Siafi','N¬∫ Termo de Compromisso'], axis=1)
df_reconstrucao_recurso = df_reconstrucao_recurso.rename(columns={'IBGE': 'ibge7'})


df_reconstrucao_recurso.head()
```

```python
df_reconstrucao_recurso.info()
```

```python
df_reconstrucao_recurso['Ano'].value_counts()
```

```python
df_reconstrucao_recurso.tail()

# Converter as colunas para datetime
df_reconstrucao_recurso['Data Envio'] = pd.to_datetime(df_reconstrucao_recurso['Data Envio'], format='%d/%m/%Y', errors='coerce')
df_reconstrucao_recurso['Data Empenho'] = pd.to_datetime(df_reconstrucao_recurso['Data Empenho'], format='%d/%m/%Y', errors='coerce')
df_reconstrucao_recurso['Data Dou'] = pd.to_datetime(df_reconstrucao_recurso['Data Dou'], format='%d/%m/%Y', errors='coerce')

# Calcular a diferen√ßa
df_reconstrucao_recurso['tempo_medio_analise'] = df_reconstrucao_recurso['Data Empenho'] - df_reconstrucao_recurso['Data Envio']
df_reconstrucao_recurso['tempo_medio_liberacao'] = df_reconstrucao_recurso['Data Dou'] - df_reconstrucao_recurso['Data Empenho']

df_reconstrucao_recurso[['Data Envio', 'Data Empenho', 'Data Dou','tempo_medio_analise', 'tempo_medio_liberacao']].head()
```

#### BASE Reconstru√ß√£o tempo m√©dio (analise e libera√ß√£o)  e contagem (por munic√≠pio)

```python
df_reconstrucao_recurso.to_csv(
    "dados/_base_final/reconstrucao_mun_completo.csv",
    index=False,           # N√£o salva o √≠ndice
    encoding='utf-8',      # Codifica√ß√£o para caracteres especiais
    sep=',',               # Separador (v√≠rgula)
    decimal='.'            # Separador decimal (ponto)
)
```

```python
df_reconstrucao_mun = df_reconstrucao_recurso.groupby('ibge7').agg({
    'tempo_medio_analise': ['mean', 'count'],
    'tempo_medio_liberacao': ['mean', 'count']
}).reset_index()

# Ajustar nomes das colunas
df_reconstrucao_mun.columns = [
    'ibge7', 
    'tempo_medio_analise', 'qtd_analise',
    'tempo_medio_liberacao', 'qtd'
]
df_reconstrucao_mun = df_reconstrucao_mun.drop(['qtd_analise'], axis=1)

df_reconstrucao_mun.head()
```

```python
df_reconstrucao_mun.to_csv(
    "dados/_base_final/reconstrucao_municipio.csv",
    index=False,           # N√£o salva o √≠ndice
    encoding='utf-8',      # Codifica√ß√£o para caracteres especiais
    sep=',',               # Separador (v√≠rgula)
    decimal='.'            # Separador decimal (ponto)
)
```
